```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
```

```python
conn = sqlite3.connect("D:/News/news.db")
conn2 = sqlite3.connect("D:/News/price.db")
```

```python
cot=pd.read_sql('SELECT * FROM COT', conn)
```

```python
product=cot[cot['PRODUCT']=='EUR']
```

```python
price = pd.read_sql("Select * from PRICE where Sym='EURUSD'",conn2)
```

```python
price.drop(['Sym'], axis=1, inplace=True)
```

```python
price['Date Time'] = price.Date + ' ' + price.Time
```

```python
price['Date Time']=pd.to_datetime(price['Date Time'])
```

```python
price.set_index('Date Time', inplace=True)
```

```python
product.set_index(pd.to_datetime(product.DATE), inplace=True)
```

```python
merged=product.join(price).dropna()
```

```python
merged
```

```python
merged['Close_1'] = merged.Close.shift()
merged['pov']=(merged.OI*np.sign(merged.Close-merged.Close_1)).cumsum()+(merged.OI*(merged.Close-merged.Close_1)/(merged[['Close_1','High']].max(axis=1)-merged[['Close_1','Low']].min(axis=1))).cumsum()
```
