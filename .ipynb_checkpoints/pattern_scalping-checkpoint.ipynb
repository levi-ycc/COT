```python
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
import sqlite3
import pandas as pd
price = pd.read_csv('D:/EURUSD.csv', names=['Date','Time','Open','High','Low','Close','Volume'])
```

```python
price.set_index(pd.to_datetime(price.Date+' '+price.Time), inplace=True)
```

```python
df=price.loc['2018-01-01':, ['Open', 'High', 'Low', 'Close']]
```

```python
df['l']=(df.High-df.Low)*1e5
df['b']=(df.Close-df.Open).abs()*1e5
df['d']=0
df.loc[df.Close>df.Open, 'd']=1
df.loc[df.Close<df.Open, 'd']=-1
```

```python
def coding(x, t):
    if x<t[0]:
        return 1
    elif x>=t[0] and x<t[1]:
        return 2
    else: return 3
```

```python
df['ld']=df.l.rolling(15).quantile(.25)
df['lu']=df.l.rolling(15).quantile(.75)
df['bd']=df.l.rolling(15).quantile(.25)
df['bu']=df.l.rolling(15).quantile(.75)
```

```python
df.dropna(inplace=True)
```

```python
df['code_l']=2
df['code_b']=2
df.loc[df.l<df.ld, 'code_l']=1
df.loc[df.l>=df.lu, 'code_l']=3
df.loc[df.b<df.bd, 'code_b']=1
df.loc[df.b>=df.bu, 'code_b']=3

df['code_l2']=df.code_l.shift()
df['code_b2']=df.code_b.shift()
df['d2']=df.d.shift()
df['l3']=df.code_l2.shift()
df['b3']=df.code_b2.shift()
df['d3']=df.d2.shift()
```

```python
signal = df.copy()
signal['signal']=0

signal.loc[(signal.code_l==3)&(signal.d==1), 'signal']=-1
# signal.loc[(signal.code_l==3)&(signal.code_d==-1), 'signal']=-1

signal.signal.sum()
```

```python
for i in range(1,11):
    pnl=(signal.signal*1e5*(signal.Close.shift(-i)-signal.Close))
    avg=pnl.sum()/signal.signal.abs().sum()
    pnl.cumsum().plot()
    plt.title(f'{i}_{round(pnl.sum(),5)}_{round(avg,5)}')
    plt.show()
```
